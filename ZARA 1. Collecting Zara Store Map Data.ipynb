```python
raw = pd.read_json('./inditex-en-el-mundo.json')
raw.head()
```

```python
location = raw['st_itxloc'].str.split(',', expand = True)
raw['latitude'] = location[0]
raw['longitude'] = location[1]
raw.head()
```

```python
data = []

for i in raw.index:
    Name = raw.loc[i, 'nomTi'],
    Chain = raw.loc[i, 'nomCad']
    Street = raw.loc[i, 'calle'] 
    City = raw.loc[i, 'pob'] 
    Region = raw.loc[i, 'reg']
    Section = raw.loc[i, 'secSen'] + ', ' + raw.loc[i, 'secCab'] + ', ' + raw.loc[i, 'secNin']
    C_c = raw.loc[i, 'centCom']
    Latitude = raw.loc[i, 'latitude']
    Longitude = raw.loc[i, 'longitude']    
    
    data.append([Name, Chain, Street, City, Region, Section, C_c, Latitude, Longitude])
    
column = ['Name', 'Chain', 'Street', 'City', 'Region', 'Section', 'C_c', 'Latitude', 'Longitude']
zara = pd.DataFrame(data, columns = column)
zara.head()
```

```python
zara.to_csv('./zara.csv', index = False)
```
