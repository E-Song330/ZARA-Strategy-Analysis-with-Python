```python
import pandas as pd
import folium, json
```

```python
reza = pd.read_csv('./reza.csv')
reza.head()
```

### Visualizing store map

```python
#Spain ['40.2085', '-3.7130']
m = folium.Map(location = ['40.2085', '-3.7130'], zoom_start = 6, tiles = 'OpenStreetMap' )
```

```python
for i in reza.index:
    lat = reza.loc[i, 'Latitude']
    lng = reza.loc[i, 'Longitude']
    name = reza.loc[i, 'Chain']
    
    
    folium.CircleMarker(location = [lat,lng],
                        tooltip = f'{name}',
                        fill = True,
                        fill_color = 'white',
                        fill_opacity = 1,
                        color = 'black',
                        weight = 1,
                        radius = 5).add_to(m)
    
#m
```

```python
from folium import FeatureGroup, LayerControl
```

```python
zara_map = folium.Map(location = ['40.2085', '-3.7130'], zoom_start = 6, tiles = 'OpenStreetMap' )

kiddy = folium.FeatureGroup(name = "Kiddy's class")
zara = folium.FeatureGroup(name = 'Zara')
lefties = folium.FeatureGroup(name = 'lefties')




for i in reza.index:
    lat = reza.loc[i, 'Latitude']
    lng = reza.loc[i, 'Longitude']
    chain = reza.loc[i, 'Chain']
    
    if chain == 'zara':
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = 'red',
                        color = 'white',
                        weight = 1,
                        radius = 5,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(zara)

    elif chain == "kiddy's class":
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = '#87ceeb',
                        color = 'white',
                        weight = 1,
                        radius = 5,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(kiddy)

    else:
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = 'white',
                        color = 'black',
                        radius = 5,
                        weight = 1,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(lefties)

kiddy.add_to(zara_map)        
zara.add_to(zara_map)   
lefties.add_to(zara_map)

folium.LayerControl(collapsed=False).add_to(zara_map)    

#zara_map
```

### Visualizing store map with population informations

```python
pop = pd.read_csv('./population.csv')
pop.head()
```

```python
pivpop = pd.read_csv('./piv_population.csv')
pivpop.head()
```

```python
geopath = './spain_provinces.geojson'
geo = json.load(open(geopath))


```

```python
geo_pop = folium.Map(location = ['40.2085', '-3.7130'], zoom_start = 6, tiles = 'cartodbpositron' )



folium.Choropleth(geo_data=geo, 
                  name='choropleth',
                  data=pivpop, 
                  columns=['PROVINCIA',"POBLACION_MUNI"],
                  key_on="feature.properties.name",
                  fill_color='YlGn',
                  fill_opacity=0.6,
                  line_opacity=0.5).add_to(geo_pop)


folium.LayerControl().add_to(geo_pop)


#geo_pop
```

```python
geo_zara = folium.Map(location = ['40.2085', '-3.7130'], zoom_start = 6, tiles = 'cartodbpositron' )


folium.Choropleth(geo_data=geo, 
                  name='choropleth',
                  data=pivpop, 
                  columns=['PROVINCIA',"POBLACION_MUNI"],
                  key_on="feature.properties.name",
                  fill_color='YlGn',
                  fill_opacity=0.6,
                  line_opacity=0.5).add_to(geo_zara)


kiddy = folium.FeatureGroup(name = "Kiddy's class")
zara = folium.FeatureGroup(name = 'Zara')
lefties = folium.FeatureGroup(name = 'lefties')


for i in reza.index:
    lat = reza.loc[i, 'Latitude']
    lng = reza.loc[i, 'Longitude']
    chain = reza.loc[i, 'Chain']
    
    if chain == 'zara':
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = 'red',
                        color = 'white',
                        weight = 1,
                        radius = 5,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(zara)

    elif chain == "kiddy's class":
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = '#87ceeb',
                        color = 'white',
                        weight = 1,
                        radius = 5,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(kiddy)

    else:
        folium.CircleMarker(location = [lat,lng], 
                        fill = True, 
                        fill_color = 'white',
                        color = 'black',
                        radius = 5,
                        weight = 1,
                        fill_opacity = 1,
                        tooltip = f'{chain}').add_to(lefties)

kiddy.add_to(geo_zara)        
zara.add_to(geo_zara)   
lefties.add_to(geo_zara)

folium.LayerControl(collapsed=False).add_to(geo_zara)

#geo_zara
```
