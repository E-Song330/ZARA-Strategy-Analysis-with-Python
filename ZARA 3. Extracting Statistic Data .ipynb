Dataset1. Population of Spain counted by pronvince

https://www.ign.es/web/ign/portal/rcc-nomenclator-nacional

```python
import pandas as pd
raw = pd.read_csv('./MUNICIPIOS.csv', sep=';', encoding='latin1')
raw.head()
```

```python
raw.columns
```

```python
pop = raw.loc[:,['PROVINCIA', 'NOMBRE_ACTUAL', 'POBLACION_MUNI', 'LONGITUD_ETRS89', 'LATITUD_ETRS89' ]]
pop.head()
```

```python
pro = pop['PROVINCIA'].replace(['Illes Balears','Araba/Álava','Alacant/Alicante','Castelló/Castellón','València/Valencia'],['Baleares','Álava','Alicante','Castellón','Valencia'])
pro
```

```python
pre = pro.reset_index()
pop['PROVINCIA'] = pre['PROVINCIA']
```

```python
pop
```

```python
pivpop = pop.pivot_table(index = {'PROVINCIA'}, values = 'POBLACION_MUNI', aggfunc = 'sum')
pivpop.head()
```

Dataset2. GDP data by province

https://www.ine.es/jaxi/Datos.htm?path=/t35/p010/rev19/&file=02001.px

```python
pibraw = pd.read_csv('./pib.csv', sep=';', thousands = '.', encoding='latin1')
pibraw.head()
```

```python
pib =pd.DataFrame()
pib['province'] = pibraw['Provincias']
```

```python
pib['PIB de Mercado'] = pibraw['Total']
```

```python
pib.head()
```

```python
PIB = pib['province'].replace(['Balears, Illes','Araba/Álava','Alicante/Alacant','Castellón/Castelló','Valencia/València','Coruña, A','Palmas, Las','Rioja, La']
                              ,['Baleares','Álava','Alicante','Castellón','Valencia', 'A Coruña','Las Palmas', 'La Rioja'])
```

```python
pib_data = PIB.reset_index()
pib['province'] = pib_data['province']
```

```python
pib.head()
```

```python
pib_pob = pd.merge(pib,pivpop, how = 'left', left_on='province', right_on='PROVINCIA')
pib_pob.head()
```

```python
pop.to_csv('./population.csv', index = False)
```

```python
pivpop.to_csv('./piv_population.csv')
```

```python
pib.to_csv('./pibdata.csv', index = False)
pib_pob.to_csv('./pib_pob.csv', index = False)
```
